```python
#WordNet

#helper function to convert treebank to wn tags
def get_wordnet_pos(treebank_tag):

    if treebank_tag.startswith('J'):
        return wn.ADJ
    elif treebank_tag.startswith('V'):
        return wn.VERB
    elif treebank_tag.startswith('N'):
        return wn.NOUN
    elif treebank_tag.startswith('R'):
        return wn.ADV
    else:
        return ''

#helper function to check for a dot
def containsPunct(word):
  charList = word.split()

  for value in charList:
    for char in value:
      if char == ".":
        return False

  return True

#helper function to check for punctuation
def containsPunctAll(word):
  charList = word.split()

  for value in charList:
    for char in value:
      if char in string.punctuation:
        return False

  return True
     



#this function takes in two review indexes , a movie index and a list
def wordNet(movieList , movieNum , reviewnumOne , reviewnumTwo):

  #this needs to be added along with the import
  lemmatizer = WordNetLemmatizer()

  allTokensTrimmed = []

  #contains best possible word sense with lesk
  senseListOne = []
  #lemmatized list
  lemmatizedOne = []
  #nonstopwords tagged
  nonStopTagged = []
  #nonPunct tagged
  nonPunctTagged = []

  senseListOneNOTNULL = []

 
  
  #full review of the selected index
  sentenceOne = movieList[movieNum]['Review'][reviewnumOne]
  #lowercase version
  tokenizedSent = nltk.word_tokenize(sentenceOne.lower())
  firstSentence = []

  #trying to lower the review to the first sentence, where if a "." is spotted
  #the appension will stop
  for i in tokenizedSent:
    if containsPunct(i):
      firstSentence.append(i)

    else:
      break

  #taggedSentence to check parts of speech before further trimming the list  
  taggedSentence = nltk.pos_tag(firstSentence)


  
  #removing stopwords from tagged list
  for word in taggedSentence:
    if word[0] not in stopwords.words('english'):
      nonStopTagged.append(word)
  
  #removing punctuation from taggedlist
  for word in nonStopTagged:
    if containsPunctAll(word[0]):
      nonPunctTagged.append(word)
      
        


  #lemmatizing the sentences
  for word in tokenizedSent:
    lemmatizedOne.append(lemmatizer.lemmatize(word))

  #creating synsets for each word using lesk to try and estimate the most probable
  #semantic meaning for the word
  for word in nonPunctTagged:
    senseListOne.append(lesk(tokenizedSent, word[0] , get_wordnet_pos(word[1])))

  #if there are None values , I want to trim them
  for word in senseListOne:
    if word is not None:
      senseListOneNOTNULL.append(word)



  

  print("Sense one " , senseListOneNOTNULL)


  
  #--- for the 2nd review---------------------------------------------------------------------------------------
  #REPEATING EVERY STEP ON THE FIRST PART HERE
  #I could have put these in a function...
  
  senseListTwo = []
  #lemmatized list
  lemmatizedTwo = []
  #nonstopwords tagged
  nonStopTaggedTwo = []
  #nonPunct tagged
  nonPunctTaggedTwo = []

  senseListTwoNOTNULL = []
  

  sentenceTwo= movieList[movieNum]['Review'][reviewnumTwo]
  #lowercase version
  tokenizedSentTwo = nltk.word_tokenize(sentenceTwo.lower())
  secondSentence = []

  #trying to lower the review to the first sentence, where if a "." is spotted
  #the appension will stop
  for i in tokenizedSentTwo:
    if containsPunct(i):
      secondSentence.append(i)

    else:
      break

  #use these for tags to feed lesk     
  taggedSentenceTwo = nltk.pos_tag(secondSentence)



  #removing stopwords from tagged list
  for word in taggedSentenceTwo:
    if word[0] not in stopwords.words('english'):
      nonStopTaggedTwo.append(word)
  
  #removing punctuation from taggedlist
  for word in nonStopTaggedTwo:
    if containsPunctAll(word[0]):
      nonPunctTaggedTwo.append(word)
      
        



  for word in tokenizedSentTwo:
    lemmatizedTwo.append(lemmatizer.lemmatize(word))


  for word in nonPunctTaggedTwo:
    senseListTwo.append(lesk(tokenizedSentTwo, word[0] , get_wordnet_pos(word[1])))

  for word in senseListTwo:
    if word is not None:
      senseListTwoNOTNULL.append(word)



  print("Sense Two " , senseListTwoNOTNULL)

  #estimating similarity by checking through both synsets with wup_similarity
  holderList = []
  highestHolder = []
  averageCount = 0
  total = 0
  #for each synset , I take the wup_similarity of that word and each of the other words
  #in the other sentence, after appending all of these to a list, I take the max value
  #of that list to get the best match for the given word
  for i in senseListOneNOTNULL:

    for j in senseListTwoNOTNULL:       
        if wn.wup_similarity(i , j) is not None:      
          holderList.append(wn.wup_similarity(i , j))
        

    averageCount += 1
    highestHolder.append(max(holderList))
    holderList.clear()

  #In the end , when all max values have been calculated, I add them together
  #and divide it by the number of total checks that have been done for each individiual
  #word that has not returned a None value , thus getting an average
  for i in highestHolder:
    total += i


  avgSimilarity = total / averageCount

  #printing the similarity rating out of 100
  print("The similarity rating is %" , avgSimilarity * 100)


  return allTokensTrimmed

#inputting multikeys , movie index , review index one , review index two
lastList = wordNet(multikeys , 0,  0 , 3)



#test...
#Most reviews will tell you what's so great about this movie and why it's worth watching, but I figured you should hear the other side of the story.
#The second entry in the Hobbit franchise improves on everything the first film got wrong.





```
